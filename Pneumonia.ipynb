```python
import os

PATH = "F://dataset//chest_xray//"
```

```python
#Counting the no of photos in train set
for file in os.listdir(PATH + "train"):
    count = 0
    for f in os.listdir(PATH + "train//" + file):
        if f.split('.')[-1] == 'jpeg':
                count = count+1
    print("Number of photos in " + file + " of train set is: {}".format(count))
```

```python
#Counting the no of photos in test set
for file in os.listdir(PATH + "test"):
    count = 0
    for f in os.listdir(PATH + "test//" + file):
        if f.split('.')[-1] == 'jpeg':
                count = count+1
    print("Number of photos in " + file + " of test set is: {}".format(count))
```

```python
#Counting the no of photos in val set
for file in os.listdir(PATH + "val"):
    count = 0
    for f in os.listdir(PATH + "val//" + file):
        if f.split('.')[-1] == 'jpeg':
                count = count+1
    print("Number of photos in " + file + " of validation set is: {}".format(count))
```
